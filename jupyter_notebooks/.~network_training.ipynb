```python
%reload_ext autoreload
%autoreload 2
%matplotlib inline


from mnist import mnist
from ldl.algorithms import *
from ldl.predictions import *
from ldl.network import Network
from ldl.utitlities import *
import numpy as np
```

```python
# With help from https://github.com/hsjeong5/MNIST-for-Numpy
#mnist.init()
```

### Get Data

```python
x_train, t_train, x_test, t_test = mnist.load()
```

```python
train_data = normalize_2d_array(x_train[:40000])
val_data = normalize_2d_array(x_train[40001:])
test_data = normalize_2d_array(x_test)

train_labels = t_train[:40000]
val_labels = t_train[40001:]
test_labels = t_test

# Convert labels to targets for output neurons
train_targets = np.zeros([train_labels.shape[0], 10])
for index, label in enumerate(train_labels):
    train_targets[index, label] = 1

val_targets = np.zeros([val_labels.shape[0], 10])
for index, label in enumerate(val_labels):
    val_targets[index, label] = 1
    
test_targets = np.zeros([test_labels.shape[0], 10])
for index, label in enumerate(test_labels):
    test_targets[index, label] = 1

```

### Set Defaults

```python

shape = [784, 100, 10]
biases = get_relu_biases(shape)
weights = get_relu_weights(shape)
epochs=10000
```

### 0.03 learning rate

```python

learning_rate = 0.03
name = '784x100x10, Learning Rate 0.03'
network = Network(weights=weights, biases=biases, name=name)
three_neg_two_by_1000 = network.train_and_validate(epochs, train_data, train_targets, val_data, val_targets, test_data, 
                                                   test_targets, test_labels, learning_rate)


```

### 0.3 learning rate

```python
learning_rate = 0.3
name = '784x100x10, Learning Rate 0.3'
network = Network(weights=weights, biases=biases, name=name)
three_neg_one_by_1000 = network.train_and_validate(epochs, train_data, train_targets, val_data, val_targets, test_data, test_targets, test_labels, learning_rate)
```

### 0.2 learning rate

```python
learning_rate = 0.2
name = '784x100x10, Learning Rate 0.2'
network = Network(weights=weights, biases=biases, name=name)
two_neg_one_by_1000 = network.train_and_validate(epochs, train_data, train_targets, val_data, val_targets, test_data, test_targets, test_labels, learning_rate)
```

### 0.1 learning rate

```python
learning_rate = 0.1
name = '784x100x10, Learning Rate 0.1'
network = Network(weights=weights, biases=biases, name=name)
one_neg_one_by_1000 = network.train_and_validate(epochs, train_data, train_targets, val_data, val_targets, test_data, test_targets, test_labels, learning_rate)
```

```python

# Combine the final results
final_results = three_neg_two_by_1000.tail(1).append(
    three_neg_one_by_1000.tail(1)
).append(
    two_neg_one_by_1000.tail(1)
).append(
    one_neg_one_by_1000.tail(1)
)
final_results
```
